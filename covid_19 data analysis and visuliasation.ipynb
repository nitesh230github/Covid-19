```python
count_death=pd.DataFrame({'desh':df.country,'mrityu':df.death})
```

```python
def change(str):
    if str == '0':
        return 0
    elif str == '1':
        return 1
    else:
        return 1
```

```python
df['death']=df['death'].apply(change)
count_death['mrityu']=count_death['mrityu'].apply(change)
```

```python
deaths=count_death.groupby('desh')['mrityu'].sum().reset_index()
df.age.fillna(df.age.mean(), inplace=True)
deaths
```

```python
df.reporting_date.fillna("1/21/2020",inplace=True)
df.reporting_date.isnull().sum()
```

```python
def dates(a):
    li = a.split('/')
    x1 = float(li[0])
    x2 = float(li[1])
    ans = (x1 * (x2 ** 2)) ** 0.5
    return ans
```

```python
df['reporting_date']= df['reporting_date'].apply(dates)
```
